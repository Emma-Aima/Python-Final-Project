````python
# 3a. Data Cleaning

import pandas as pd

# Load the dataset
df = pd.read_csv('owid-covid-data.csv')

# 1. Filter countries of interest (Kenya, USA, India)
countries = ['Kenya', 'United States', 'India']
filtered_df = df[df['location'].isin(countries)].copy()

# 2. Drop rows with missing dates or critical values
# Critical columns: date, location, total_cases, total_deaths
filtered_df = filtered_df.dropna(subset=['date', 'location', 'total_cases', 'total_deaths'])

# 3. Convert date column to datetime
filtered_df['date'] = pd.to_datetime(filtered_df['date'])

# 4. Handle missing numeric values
# For time-series data, forward fill is often appropriate for cases/deaths
numeric_cols = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 
                'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated']

# Forward fill for cumulative metrics (they never decrease)
filtered_df[numeric_cols] = filtered_df.groupby('location')[numeric_cols].ffill()

# For daily metrics like new_cases, we can fill zeros or interpolate
filtered_df['new_cases'] = filtered_df['new_cases'].fillna(0)
filtered_df['new_deaths'] = filtered_df['new_deaths'].fillna(0)

# Verify the cleaned data
print(f"Filtered countries: {filtered_df['location'].unique()}")
print(f"Date range: {filtered_df['date'].min()} to {filtered_df['date'].max()}")
print("\nMissing values after cleaning:")
print(filtered_df[numeric_cols].isnull().sum())

# Optional: Save cleaned data
filtered_df.to_csv('cleaned_covid_data.csv', index=False)
````

````python
#4. Exploratory Data Analysis (EDA)

import matplotlib.pyplot as plt
import seaborn as sns

# Load and prepare data
df = pd.read_csv('owid-covid-data.csv')
countries = ['Kenya', 'United States', 'India']
df = df[df['location'].isin(countries)]

# Convert date and handle missing values
df['date'] = pd.to_datetime(df['date'])
df.sort_values(['location', 'date'], inplace=True)

# Forward fill missing values for time-series continuity
cols_to_fill = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths']
df[cols_to_fill] = df.groupby('location')[cols_to_fill].ffill()

# Calculate death rate
df['death_rate'] = df['total_deaths'] / df['total_cases']

# Set up visualization style
sns.set(style="whitegrid")
plt.figure(figsize=(14, 10))

# 1. Plot total cases over time
plt.subplot(2, 2, 1)
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)
plt.title('Total COVID-19 Cases Over Time')
plt.ylabel('Total Cases')
plt.legend()

# 2. Plot total deaths over time
plt.subplot(2, 2, 2)
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)
plt.title('Total COVID-19 Deaths Over Time')
plt.ylabel('Total Deaths')
plt.legend()

# 3. Compare daily new cases (7-day rolling average for smoother visualization)
plt.subplot(2, 2, 3)
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], 
             country_data['new_cases'].rolling(7).mean(), 
             label=country)
plt.title('Daily New Cases (7-day Average)')
plt.ylabel('New Cases')
plt.legend()

# 4. Plot death rate over time
plt.subplot(2, 2, 4)
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], 
             country_data['death_rate'].rolling(7).mean(), 
             label=country)
plt.title('COVID-19 Death Rate Over Time')
plt.ylabel('Death Rate (Deaths/Cases)')
plt.legend()

plt.tight_layout()
plt.show()

# Additional visualizations
# Bar chart of total cases by country
plt.figure(figsize=(10, 6))
latest_data = df.sort_values('date').groupby('location').last()
sns.barplot(x=latest_data.index, y=latest_data['total_cases'])
plt.title('Total COVID-19 Cases by Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()

# Heatmap of correlations (for numeric columns)
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of COVID-19 Metrics')
plt.show()
````

````python
# 5. Visualizing Vaccination Progress
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and prepare data
df = pd.read_csv('owid-covid-data.csv')
countries = ['Kenya', 'United States', 'India', 'United Kingdom', 'Brazil']
df = df[df['location'].isin(countries)]

# Convert date and calculate vaccination metrics
df['date'] = pd.to_datetime(df['date'])
df.sort_values(['location', 'date'], inplace=True)

# Calculate vaccination percentages
df['pct_vaccinated'] = (df['people_vaccinated'] / df['population']) * 100
df['pct_fully_vaccinated'] = (df['people_fully_vaccinated'] / df['population']) * 100

# Fill missing values with forward fill for time-series continuity
vaccine_cols = ['total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
                'pct_vaccinated', 'pct_fully_vaccinated']
df[vaccine_cols] = df.groupby('location')[vaccine_cols].ffill()

# Set visualization style
sns.set(style="whitegrid")
plt.figure(figsize=(16, 12))

# 1. Plot cumulative vaccinations over time
plt.subplot(2, 2, 1)
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], 
             country_data['total_vaccinations']/1e6,  # Display in millions
             label=country)
plt.title('Total COVID-19 Vaccinations Over Time')
plt.ylabel('Vaccinations (Millions)')
plt.legend()

# 2. Compare % vaccinated population
plt.subplot(2, 2, 2)
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], 
             country_data['pct_vaccinated'], 
             label=f'{country} (at least 1 dose)')
    plt.plot(country_data['date'], 
             country_data['pct_fully_vaccinated'], 
             linestyle='--',
             label=f'{country} (fully vaccinated)')
plt.title('Vaccination Percentage of Population')
plt.ylabel('Percentage Vaccinated')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# 3. Latest vaccination status pie chart (for the most recent date)
plt.subplot(2, 2, 3)
latest_date = df['date'].max()
latest_data = df[df['date'] == latest_date]

for i, country in enumerate(countries):
    country_data = latest_data[latest_data['location'] == country].iloc[0]
    vaccinated = country_data['people_vaccinated']
    fully_vaccinated = country_data['people_fully_vaccinated']
    unvaccinated = country_data['population'] - vaccinated
    
    plt.subplot(2, 3, i+4)  # Adjust subplot position for pie charts
    plt.pie([vaccinated, unvaccinated],
            labels=['Vaccinated', 'Unvaccinated'],
            autopct='%1.1f%%',
            colors=['#4c72b0', '#dd8452'])
    plt.title(f'{country}\n({latest_date.date()})')

plt.suptitle('COVID-19 Vaccination Progress Analysis', y=1.02)
plt.tight_layout()
plt.show()

# Additional analysis: Vaccination rollout speed
print("\nVaccination Rollout Speed Analysis:")
for country in countries:
    country_data = df[df['location'] == country].dropna(subset=['people_fully_vaccinated'])
    if not country_data.empty:
        start_date = country_data['date'].min()
        first_vaccination = country_data['people_fully_vaccinated'].iloc[0]
        print(f"\n{country}:")
        print(f"Vaccinations started: {start_date.date()}")
        print(f"Days to reach 10% fully vaccinated: {country_data[country_data['pct_fully_vaccinated'] >= 10]['date'].iloc[0].date()}")
        print(f"Days to reach 50% fully vaccinated: {country_data[country_data['pct_fully_vaccinated'] >= 50]['date'].iloc[0].date() if any(country_data['pct_fully_vaccinated'] >= 50) else 'Not yet reached'}")
````

````python
# Import libraries and load data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('owid-covid-data.csv')
selected_countries = ['Kenya', 'United States', 'India', 'United Kingdom', 'Brazil']
covid = df[df['location'].isin(selected_countries)].copy()
````

````python
# KEY INSIGHTS
# 1. Vaccination Rollouts Disparities 

# Plot vaccination percentages
plt.figure(figsize=(12,6))
for country in selected_countries:
    country_data = covid[covid['location'] == country]
    plt.plot(country_data['date'], country_data['people_fully_vaccinated_per_hundred'], label=country)
plt.title('Full Vaccination Coverage (% Population)')
plt.legend()
plt.show()

# Insight: The UK and US achieved >50% full vaccination within 200 days of vaccine availability, 
# while Kenya took 150% longer to reach the same threshold. This highlights significant global vaccine inequality.
````

````python
# 2. Case Fatality Rate Trends 

# Calculate and plot death rates
covid['death_rate'] = covid['total_deaths']/covid['total_cases']
plt.figure(figsize=(12,6))
sns.lineplot(data=covid, x='date', y='death_rate', hue='location')
plt.title('Case Fatality Rate Over Time')
plt.show()

# Insight: Early pandemic fatality rates peaked at 3-6% across all countries but converged to 1-2% by 2022, 
# suggesting improved treatment protocols and protective effects of vaccination.
````

````python
# 3. Wave Pattern Synchronization

# Plot new case waves
plt.figure(figsize=(12,6))
for country in selected_countries:
    country_data = covid[covid['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases_smoothed_per_million'], label=country)
plt.title('Case Waves per Million People')
plt.legend()
plt.show()

# Pattern: All countries experienced synchronized waves corresponding with variant emergence 
# (Delta in Q2 2021, Omicron in Q4 2021), but with different magnitudes based on mitigation policies.
````

````python
# Anomalies and Notable Findings
# 1. India's Exceptional Second Wave

india = covid[covid['location'] == 'India']
plt.plot(india['date'], india['new_cases_smoothed_per_million'])
plt.title('India\'s Dramatic Delta Wave (April-May 2021)')
plt.show()

# Anomaly: India showed a 10x spike in daily cases during its Delta wave compared to other countries, 
# suggesting unique epidemiological factors or under-reporting in earlier waves.
````

````python
 # 2. Brazil's Delayed But Rapid Vaccination

# Vaccination timeline comparison
vaccine_start = covid.groupby('location')['date'].min()
print(vaccine_start.sort_values())

# Finding: Despite late start (3 months after US/UK), Brazil achieved faster initial rollout due to prior mass vaccination experience with other diseases.
````
